```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier


from IPython.display import Image
import graphviz
import pydotplus 

pd.options.display.max_columns = 25
```

```python
df = pd.read_csv("train.csv", index_col="PassengerId")
df['Sex'].replace(to_replace=['male', 'female'], value=[0, 1], inplace=True)
df.fillna(0, inplace=True)
```

```python
features = ["Pclass", 'Sex', 'Age', 'SibSp']
target = "Survived"
```

```python
def train_tree(state):
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], random_state=state) 
    tree_class = tree.DecisionTreeClassifier(random_state=0, max_depth=4)
    tree_class.fit(X_train, y_train)
    print("Правильность на обучающем наборе: {:.3f}".format(tree_class.score(X_train, y_train)))
    print("Правильность на тестовом наборе: {:.3f}".format(tree_class.score(X_test, y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], random_state=42) 
tree_class = tree.DecisionTreeClassifier(random_state=0, max_depth=4)
tree_class.fit(X_train, y_train)
print("Правильность на обучающем наборе: {:.3f}".format(tree_class.score(X_train, y_train)))
print("Правильность на тестовом наборе: {:.3f}".format(tree_class.score(X_test, y_test)))
```

```python
array_train = []
array_test = []
y = []
for i in range(0, 100):
    X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], random_state=i) 
    tree_class = tree.DecisionTreeClassifier(random_state=0, max_depth=4)
    tree_class.fit(X_train, y_train)
    array_train.append(tree_class.score(X_train, y_train))
    array_test.append(tree_class.score(X_test, y_test))
    y.append(i)
plt.plot(y, array_train, '-r', y, array_test, '-b')
```
