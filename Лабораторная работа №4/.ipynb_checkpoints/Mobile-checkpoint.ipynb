```python
import pandas as pd
import numpy as np

import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap, colorConverter, LinearSegmentedColormap

from sklearn.manifold import TSNE

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics.cluster import silhouette_score
```

```python
# Функция discrete_scatter из mglearn

def discrete_scatter(x1, x2, y=None, markers=None, s=10, ax=None,
                     labels=None, padding=.2, alpha=1, c=None, markeredgewidth=None):
    """Adaption of matplotlib.pyplot.scatter to plot classes or clusters.

    Parameters
    ----------

    x1 : nd-array
        input data, first axis

    x2 : nd-array
        input data, second axis

    y : nd-array
        input data, discrete labels

    cmap : colormap
        Colormap to use.

    markers : list of string
        List of markers to use, or None (which defaults to 'o').

    s : int or float
        Size of the marker

    padding : float
        Fraction of the dataset range to use for padding the axes.

    alpha : float
        Alpha value for all points.
    """
    if ax is None:
        ax = plt.gca()

    if y is None:
        y = np.zeros(len(x1))

    unique_y = np.unique(y)

    if markers is None:
        markers = ['o', '^', 'v', 'D', 's', '*', 'p', 'h', 'H', '8', '<', '>'] * 10

    if len(markers) == 1:
        markers = markers * len(unique_y)

    if labels is None:
        labels = unique_y

    # lines in the matplotlib sense, not actual lines
    lines = []

    current_cycler = mpl.rcParams['axes.prop_cycle']

    for i, (yy, cycle) in enumerate(zip(unique_y, current_cycler())):
        mask = y == yy
        # if c is none, use color cycle
        if c is None:
            color = cycle['color']
        elif len(c) > 1:
            color = c[i]
        else:
            color = c
        # use light edge for dark markers
        if np.mean(colorConverter.to_rgb(color)) < .4:
            markeredgecolor = "grey"
        else:
            markeredgecolor = "black"

        lines.append(ax.plot(x1[mask], x2[mask], markers[i], markersize=s,
                             label=labels[i], alpha=alpha, c=color,
                             markeredgewidth=markeredgewidth,
                             markeredgecolor=markeredgecolor)[0])

    if padding != 0:
        pad1 = x1.std() * padding
        pad2 = x2.std() * padding
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()
        ax.set_xlim(min(x1.min() - pad1, xlim[0]), max(x1.max() + pad1, xlim[1]))
        ax.set_ylim(min(x2.min() - pad2, ylim[0]), max(x2.max() + pad2, ylim[1]))

    return lines
```

```python
df = pd.read_csv('mobile.txt', sep='\t', encoding='1251')
```

```python
transformer = TSNE(random_state=42, n_components=2)
df_two = transformer.fit_transform(df)
```

```python
df_two
```

```python
df.drop(['Код'], axis=1, inplace=True)
```

```python
kmeans = KMeans(n_clusters=2).fit(df)
silhouette_score(df, kmeans.predict(df))
```

```python
kmeans = KMeans(n_clusters=2).fit(df_two)
silhouette_score(df_two, kmeans.predict(df_two))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# axes[0][0].set_title(
#    f"Количество кластеров: 2, силуэтный коэффициент: {silhouette_score(df_two, kmeans.predict(df_two))}")

kmeans = KMeans(n_clusters=2).fit(df_two)
axes[0][0].set_title(silhouette_score(df_two, kmeans.predict(df_two)))
discrete_scatter(df_two[:, 0], df_two[:, 1], kmeans.labels_, ax=axes[0][0])

kmeans = KMeans(n_clusters=3).fit(df_two)
axes[0][1].set_title(silhouette_score(df_two, kmeans.predict(df_two)))
discrete_scatter(df_two[:, 0], df_two[:, 1], kmeans.labels_, ax=axes[0][1])

kmeans = KMeans(n_clusters=4).fit(df_two)
axes[1][0].set_title(silhouette_score(df_two, kmeans.predict(df_two)))
discrete_scatter(df_two[:, 0], df_two[:, 1], kmeans.labels_, ax=axes[1][0])

kmeans = KMeans(n_clusters=5).fit(df_two)
axes[1][1].set_title(silhouette_score(df_two, kmeans.predict(df_two)))
discrete_scatter(df_two[:, 0], df_two[:, 1], kmeans.labels_, ax=axes[1][1])

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

agg = AgglomerativeClustering(n_clusters=2)
assignment = agg.fit_predict(df_two)
axes[0][0].set_title(silhouette_score(df_two, assignment))
discrete_scatter(df_two[:, 0], df_two[:, 1], agg.labels_, ax=axes[0][0])

agg = AgglomerativeClustering(n_clusters=3)
assignment = agg.fit_predict(df_two)
axes[0][1].set_title(silhouette_score(df_two, assignment))
discrete_scatter(df_two[:, 0], df_two[:, 1], agg.labels_, ax=axes[0][1])

agg = AgglomerativeClustering(n_clusters=4)
assignment = agg.fit_predict(df_two)
axes[1][0].set_title(silhouette_score(df_two, assignment))
discrete_scatter(df_two[:, 0], df_two[:, 1], agg.labels_, ax=axes[1][0])

agg = AgglomerativeClustering(n_clusters=5)
assignment = agg.fit_predict(df_two)
axes[1][1].set_title(silhouette_score(df_two, assignment))
discrete_scatter(df_two[:, 0], df_two[:, 1], agg.labels_, ax=axes[1][1])

plt.show()
```

```python
d = DBSCAN(eps=0.1)
clusters = d.fit_predict(df_two)
discrete_scatter(df_two[:, 0], df_two[:, 1], d.labels_)
plt.show()
print(len(set(clusters)))
print(np.unique(clusters))
print("Размеры кластеров: {}".format(np.bincount(clusters + 1)))
```

```python
scaler = StandardScaler()
scaler.fit(df_two)
df_scaled = scaler.transform(df_two)
d = DBSCAN(eps=0.1)
clusters = d.fit_predict(df_scaled)
discrete_scatter(df_scaled[:, 0], df_scaled[:, 1], d.labels_)
plt.show()
print(len(set(clusters)))
print(np.unique(clusters))
print("Размеры кластеров: {}".format(np.bincount(clusters + 1)))
```
